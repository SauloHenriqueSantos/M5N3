```python
# Importar a biblioteca pandas
import pandas as pd
import numpy as np

# Definir o caminho do arquivo CSV
caminho_arquivo = r"D:\Escritório\Desenvolvedor\Mundo05\M5N3\pico_web.csv"

# Passo 1: Ler o conteúdo do CSV, atentando para o separador, engine e encoding
dados = pd.read_csv(caminho_arquivo, sep=';', engine='python', encoding='utf-8')

# Passo 2: Verificar a importação dos dados
print("Informações gerais sobre o conjunto de dados:")
print(dados.info())
print("\nPrimeiras 5 linhas do conjunto de dados:")
print(dados.head())
print("\nÚltimas 5 linhas do conjunto de dados:")
print(dados.tail())

# Passo 3: Criar uma cópia do conjunto de dados original
dados_copia = dados.copy()

# Passo 4: Substituir valores nulos da coluna 'Calories' por 0
dados_copia['Calories'] = dados_copia['Calories'].fillna(0)
print("\nConjunto de dados após substituir valores nulos em 'Calories' por 0:")
print(dados_copia)

# Passo 5: Substituir valores nulos da coluna 'Date' por '1900/01/01'
dados_copia['Date'] = dados_copia['Date'].fillna('1900/01/01')
print("\nConjunto de dados após substituir valores nulos em 'Date' por '1900/01/01':")
print(dados_copia)

# Passo 6: Converter a coluna 'Date' para datetime
try:
    dados_copia['Date'] = pd.to_datetime(dados_copia['Date'], format='%Y/%m/%d')
except Exception as e:
    print("\nErro encontrado ao tentar converter 'Date':", e)

# Passo 7: Substituir '1900/01/01' por NaN na coluna 'Date'
dados_copia['Date'] = dados_copia['Date'].replace('1900/01/01', np.nan)

# Passo 8: Repetir a conversão de 'Date' para datetime
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'], errors='coerce')
print("\nConjunto de dados após converter 'Date' para datetime (com tratamento de erro):")
print(dados_copia)

# Passo 9: Corrigir o valor específico "20201226" na coluna 'Date'
dados_copia['Date'] = dados_copia['Date'].replace('20201226', '2020/12/26')
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'], errors='coerce')

# Passo 10: Remover registros com valores nulos
dados_copia = dados_copia.dropna(subset=['Date'])

# Exibir o DataFrame final para verificar as mudanças
print("\nDataframe final após todas as transformações:")
print(dados_copia)
```
